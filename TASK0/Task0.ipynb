```python
def merge_image(im_nr):
    r = cv2.imread(f"data/{im_nr}v_R.jpg")
    b = cv2.imread(f"data/{im_nr}v_B.jpg")
    g = cv2.imread(f"data/{im_nr}v_G.jpg")
    r = cv2.cvtColor(r, cv2.COLOR_BGR2GRAY)
    b = cv2.cvtColor(b, cv2.COLOR_BGR2GRAY)
    g = cv2.cvtColor(g, cv2.COLOR_BGR2GRAY)

    best = channels.align(r,g,b)
    utils.show_plot(best, "Group 11-13 12", f"task0_colorized_{im_nr}.png")
    
image_nrs = ['00125','00149','00351']
for nr in image_nrs:
    merge_image(nr)
```
